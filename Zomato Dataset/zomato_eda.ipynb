```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

### both have country code, so we could combine them

```python
final_df = pd.merge(df, df_country , on = 'Country Code' , how='left')
```

```python
final_df.head()
```

```python
### to check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_values= final_df.Country.value_counts().values
## checking the values on every country
country_names = final_df.Country.value_counts().index
```

```python
#piechart for top 3 countries

plt.pie(country_values[:3] , labels=country_names[:3] ,autopct='%1.2f%%');
##add this to get percentage values autopct='%1.2f%%'
```

### Observations : more zomato records are From india followed by US and UK.

```python
final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

```python
ratings.describe()
```

## Observations : 

1. when ratings are between 4.5 to 4.9 is excellent
2. when ratings are between 4.0 to 4.4 , its verygood
3. when ratings are between 3.5 to 3.9 , its good
4. when ratings are in between 3.0 to 3.4, its average
5. when ratings are in between 2.5 to 2.9 , its average
5. when ratings are in between 0,1 to 2,4 , its poor


```python
plt.rcParams['figure.figsize'] = (15,8)
sns.barplot( x = 'Aggregate rating' , y = 'Rating Count' , data = ratings)
```

## i want to get the plot on basis of colors mentioned in ratings data frame

```python
sns.barplot( x = 'Aggregate rating' , y = 'Rating Count',hue = 'Rating color' , data = ratings , palette = ['black', 'red', 'orange','yellow','green','green'])
```

## observations :
1. Not rated is the highest count
2. most ratings are coming under 2.5 to 3.9

```python
ratings.isnull().sum()
```

## if anything value misses, we could usethe values between 2.5 to 3.9

```python
## count plot for categorical data
sns.countplot( x= 'Rating color', data = ratings,  palette = ['black', 'red', 'orange','yellow','lime','darkgreen'])
```
